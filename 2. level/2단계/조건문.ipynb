```python
A,B = map(int, input().split())
C = int(input())

if B+C < 60:
    print(A, B+C)
else:
    K = A+((B+C)//60)
    if K < 23:
        print(K , (B+C)%60)
    else:
        print(K%24,  (B+C)%60)  
```
