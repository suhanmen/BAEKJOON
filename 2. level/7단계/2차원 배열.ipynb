```python
N,M = map(int, input().split())

a = []
b = []
total = []

for _ in range(N):
    a.append(list(map(int,input().split())))
    total.append([])
    
for _ in range(N):
    b.append(list(map(int,input().split())))
    
for i in range(N):
    for j in range(M):   
        total[i].append(a[i][j]+b[i][j])

for i in range(N):
    print(*total[i])
```

# 4 - 다시풀기
num = int(input())   # 색종이의 수  
paper = [[0 for _ in range(101)]for _ in range(101)]    # 2차원 배열  

for _ in range(num):   
    x, y = map(int, input().split())  
    for i in range(x, x+10):                            # 가로가 10cm. 즉, 10칸  
        for j in range(y, y+10):                        # 세로가 10cm. 즉, 10칸  
            paper[i][j] = 1                             # 가로, 세로에 해당하는 사각형의 범위에 1 채워넣기.  

ans = 0  
for r in paper:                                           # 배열을 가져오기
    ans += r.count(1)                                     # 1이 존재하는 갯수 세어보기
print(ans)
