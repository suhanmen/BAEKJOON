# 2

```python
T = int(input())
total = []

for _ in range(T):
    total.append(list(map(int, input().split())))
    
for i in range(T):
    print(sum(total[i]))
```

# 3

```python
n = int(input())
total = 0
for i in range(1, n+1):
    total += i
    
print(total)
```

# 4

```python
X = int(input())
N = int(input())
t = []
total = 0

for _ in range(N):
    t.append(list(map(int, input().split())))
for i in range(N):
    total += t[i][0]*t[i][1]

print('Yes') if X == total else print('No')
```
