```python
N = int(input())

t=1
for i in range(1,N+1):
    print(' ' * (N-i) + ('*' * t))
    t += 2

t -= 2
for i in range(N,((2*N)-1)):
    t -= 2
    print(' ' * (i-(N-1)) + ('*' * t))
```

# 4
N, M = map(int, input().split())  
B = [i for i in range(1, N + 1)]  
for _ in range(M):
    i, j, k = map(int, input().split())  
    B[i-1:j] = B[k-1:j] + B[i-1:k-1]  
print(*B)  

```python
N, M= map(int, input().split())

t = []
for i in range(N+1):
    t.append(i)
      
for _ in range(M):
    i,j,k = map(int,input().split())
    
    p= t[i:k]

    a = len(t[k:j+1])
    b = len(t[i:k])

    t[i : i+a] = t[k:j+1]
    t[i+a :i+a+b] = p


print(*t[1:])
```

```python
C = int(input())

for i in range(C):
    a, *b = map(int, input().split())

    
    m = sum(b)/ a
    k = 0
    
    for j in b:
        if m < j:
            k += 1
            
    total = (k/a)*100
    print('{:.3f}%'.format(total))
```

    word = word.replace(i, '*')  # input 변수와 동일한 이름의 변수

# 8 다시보기
result = 0  
for i in range(int(input())):  
    word = input()  
    if list(word) == sorted(word, key=word.find):  
        result += 1  
print(result)

```python
t = 0

for _ in range(int(input())):
    k = input()
    
    if len(k) == 1 or len(k) == 2:
        t += 1
   
    else:
        for i in range(len(k)-2):
            if k[i] == k[i+2]:
                break
                
            elif i == (len(k)-3):
                t += 1
            
print(t)
```

```python
result = 0  
for i in range(int(input())):  
    word = input()
    print(list(word))
    print(sorted(word, key=word.find))
    if list(word) == sorted(word, key=word.find):  
        result += 1  

print(result)
```
