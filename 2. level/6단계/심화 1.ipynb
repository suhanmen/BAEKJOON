```python
N = int(input())

t=1
for i in range(1,N+1):
    print(' ' * (N-i) + ('*' * t))
    t += 2

t -= 2
for i in range(N,((2*N)-1)):
    t -= 2
    print(' ' * (i-(N-1)) + ('*' * t))
```

```python
N, M= map(int, input().split())

t = []
for i in range(N):
    t.append(i+1)
    
    
for _ in range(M):
    i,j,k = map(int,input().split())
    
    p= t[i-1:k]
    print(len(t[i-1:k]),len(t[k-1:j]))
    
    if len(t[i-1:k]) == len(t[k-1:j]):
        
        t[i-1:k] = t[k-1:j]
        t[k-1:j] = p
    else:
        print('error')
    
    print(t)
    
```
