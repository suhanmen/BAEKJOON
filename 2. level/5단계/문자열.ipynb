```python
n = input()
di = {}
alpa = 'abcdefghigjklmnopqrstuvwxyz'
for i in alpa:
    di[i] = -1

t = 0
for i in n:
    if di[i] != -1:
        t += 1
        continue
    
    di[i] = t
    t += 1

m = list(di.values())

for i in range(len(di)):
    print(m[i], end = ' ')
```

```python
n = input().upper()
d = {}
u = 0

for i in n:
    if i in d.keys():
        d[i] += 1
    
    else:
        d[i] = 1
        
    
    if u < d[i]:
        k = i
        u = d[i]
    elif u == d[i]:
        k = '?'

print(k)
```
