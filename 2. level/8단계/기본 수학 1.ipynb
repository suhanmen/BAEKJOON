```python
from fractions import Fraction

X = int(input())
a = []
k = 0
total = 0
n = 2

while total <= X:
    p = []
    a.append([])
    for j in range(1,n):
        p.append(j)
    for i in range(len(p)):
        a[k].append(p[i]/p[-i-1])    
        total += 1
    k += 1
    n += 1
    
point = 0
for i in range(len(a)):
    if i % 2 == 0:
        a[i] = a[i][::-1]
    if i < len(a):
        point += len(a[i])

print(Fraction(a[len(a)-1][X -point -1]))
```
