a = int(input())  
 
num = 1 # 최소 시작이 1이기 때문  
num_six = 6 # 육각형의 벌집 구조는 6씩 증가  
cnt = 1 # 이동하기 위해 거쳐야 하는 방의 수, 최솟값이 1  
 
while a > num: # 입력받은 a가 num 보다 작은 동안 반복  
    num += num_six # 시작 1부터 차례대로 num_six를 더함  
    num_six += 6 # 6, 12, 18... 로 차례대로 늘어남  
    cnt += 1 # 거쳐야 하는 방의 수  
     
print(cnt) # 거쳐야 하는 방의 수 출력

1개 1 -> 1\~1

2개 6 -> 2\~7

3개 12 -> 8\~19       

4개 18 -> 20\~37     

5개 24 -> 38\~61   

```python
n = int(input())

b = 1

k = 1
i = 0

if n == 1:
    b = 1
else:
    while n > k + 6*i:
        b += 1
        k = k+6*i
        i += 1
    
print(b)
```
