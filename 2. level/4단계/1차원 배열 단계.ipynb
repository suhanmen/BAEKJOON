# 5

```python
N,M = map(int, input().split())

total = []
for _ in range(N):
    total.append([0])
    
    
for t in range(M):
    i,j,k = map(int, input().split())
    
    for p in range(i-1,j):
            total[p] = [k]

for i in range(N):
    print(total[i][0], end = ' ')
```

# 6

```python
N,M = map(int, input().split())

total = []
for i in range(N):
    total.append([i+1])
    
    
for t in range(M):
    i,j = map(int, input().split())
    i = i-1
    j = j-1
    
    number = total[i]
    total[i] = total[j]
    total[j] = number

for i in range(N):
    print(total[i][0], end = ' ')
```

# 9

```python
N,M = map(int, input().split())

total = []
for i in range(N):
    total.append([i+1])
    
    
for _ in range(M):
    i,j = map(int, input().split())
    p = []
    for k in range(j-1,i-2, -1):
        p.append(total[k])
    
    o = 0
    for k in range(i-1,j):
        total[k] = p[o]
        o += 1

for i in range(N):
    print(total[i][0], end = ' ')      
```

# 10

```python
N = int(input())
test = list(map(int, input().split()))
m = max(test)


for i in range(N):
    if test[i] != m:
        test[i] = float(test[i] / m * 100)

print(test)
print(sum(test))
```
